<b>Hypothesis 1</b>
Which fb page total interaction was greater than other?

The below bar graph shows that the total interaction in FB Page for Right Winged news is greater and highest than other two pages i.e. Daily Vine & Silence is Consent

```python
from itertools import cycle, islice
my_colors = list(islice(cycle(['b', 'g', 'r']), None, len(df)))
SumOfTotalInteraction = df.groupby('Page Name')['Total Interactions'].sum().reset_index(name='Tot Int')
print(SumOfTotalInteraction)

x = SumOfTotalInteraction['Page Name']
y = SumOfTotalInteraction['Tot Int']
plt.bar(x , y, color=my_colors)
plt.title("   Total Interaction by Pages from 2011-2018")
plt.xticks(rotation=30, horizontalalignment="center")
plt.xlabel("Page Name")
plt.ylabel("Sum of Total Interaction in billion")
plt.show()
# plot.show(block=True);
```

But, the below Bar graph shows that the page "Silence is Consent" had highest video views comparatively to others

```python

SumOfFBVideosViews = df.groupby('Page Name')['Owned Video Views (FB Video)'].sum()
SumOfOwnVideosViews=df.groupby('Page Name')['Owned Video Views from Page Posts (FB Video and Live)'].sum()
SumOfOwnfbVideosViews=df.groupby('Page Name')['Owned Video Views (FB Video)'].sum()
SumOfAllVideo= (SumOfFBVideosViews+ SumOfOwnVideosViews + SumOfOwnfbVideosViews).reset_index(name='Tot video view')
print(SumOfAllVideo)
x = SumOfAllVideo['Page Name']
y = SumOfAllVideo['Tot video view']
plt.xticks(rotation=30, horizontalalignment="center")
plt.xlabel("Page Name")
plt.ylabel("Sum of FbVideos & OwnedVideos in Millions")
plt.bar(x , y, color=my_colors)
plt.title("    Total Video Views by Pages from 2011 -2018")
plt.show()
```

<b>Hypothesis 2</b>
 
Correlation between the Facebook live interaction and Page likes?

```python
#Correlation
colum1=df['Facebook Live Interactions']
column2=df['Page Likes']
correlation=colum1.corr(column2)
print(correlation)
```

```python
import seaborn as sns

sns.regplot(x=df["Facebook Live Interactions"], y=df["Page Likes"])
```

This Correlation is positive correlation. It is partially Correlated. The Facebook live interaction is partially proportional to the Page Likes. More the Facebook Interaction , the Page like is increasing Slightly
